```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import datetime
```

```python
# Load Infosys dataset
file_path = r"C:\Users\91942\Downloads\Infosys.xlsx"  # Adjust file path if needed
df = pd.read_excel(file_path)

df.head(26)
```

```python
print(df.head())

# Convert 'Date' column to datetime format
df['Date '] = pd.to_datetime(df['Date '])

# Set 'Date' as the index
df.set_index('Date ', inplace=True)
```

```python
# --------------------- 1. Daily Price Fluctuations and Market Events -------------------------
# Calculate daily price changes
df['Daily Change'] = df['close '].pct_change()

# Plot daily price fluctuations
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Daily Change'], color='blue', label='Daily Price Fluctuation')
plt.axhline(y=0, color='gray', linestyle='--')
plt.title('Daily Price Fluctuations in Infosys Stock')
plt.xlabel('Date')
plt.ylabel('Daily Change (%)')
plt.legend()
plt.grid()
plt.show()
```

```python
# --------------------- 2. Trading Volume Impact on Price Stability -------------------------
# Correlation between Trading Volume and Close Price
correlation = df['VOLUME '].corr(df['close '])
print(f"Correlation between Trading Volume and Close Price: {correlation:.2f}")

# Scatter plot for Volume vs Close Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='VOLUME ', y='close ', data=df, color='green')
plt.title('Trading Volume vs. Stock Price')
plt.xlabel('Trading Volume')
plt.ylabel('Stock Price (Close)')
plt.grid()
plt.show()
```

```python
# --------------------- 3. Significant Market Events and Price Fluctuations -------------------------
# Highlight significant price drop/spike (e.g., 2% threshold for this example)
significant_events = df[df['Daily Change'].abs() > 0.02]

# Plot close prices with significant fluctuations highlighted
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['close '], label='Stock Price (Close)', color='blue')
plt.scatter(significant_events.index, significant_events['close '], color='red', label='Significant Events', zorder=3)
plt.title('Significant Market Events and Price Fluctuations')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.grid()
plt.show()
```
